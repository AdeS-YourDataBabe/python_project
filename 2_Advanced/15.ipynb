```python
# importing libraries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# loading datasets
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# cleaning dataset
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = (df['job_skills']
                    .apply(lambda x: ast.literal_eval(x) 
                           if pd.notna(x) else x))
```

```python
df
```

```python
df_DA = df[df['job_title_short'] == 'Data Analyst'].copy()
```

```python
df_DA['job_posted_month_num'] = df_DA['job_posted_date'].dt.month

df_DA_explode = df_DA.explode('job_skills')

df_DA_pivot = df_DA_explode.pivot_table(index = 'job_posted_month_num', columns= 'job_skills', aggfunc = 'size', fill_value = 0)

df_DA_pivot.loc['Total'] = df_DA_pivot.sum()

# df_DS_pivot[df_DS_pivot.loc['Total'].sort_values(ascending=False).head().index]

df_DA_pivot = (df_DA_pivot[df_DA_pivot.loc['Total']
                           .sort_values(ascending=False)
                           .head().index])

df_DA_pivot = df_DA_pivot.drop('Total')

df_DA_pivot = df_DA_pivot.reset_index()

df_DA_pivot['job_posted_month'] = (df_DA_pivot['job_posted_month_num']
                                   .apply(lambda x: 
                                          pd.to_datetime(x, format ='%m')
                                          .strftime('%b')))

df_DA_pivot = df_DA_pivot.set_index('job_posted_month')

df_DA_pivot = df_DA_pivot.drop(columns='job_posted_month_num')

df_DA_pivot.plot(
    kind='line',
    linewidth = 4, 
    linestyle = ':', 
    colormap='plasma',
    marker = 'o',
    markersize = 6,
    figsize=(10,5)
)

plt.title('Top 5 Skills for Data Analysts Monthly')
plt.ylabel('Count')
plt.xlabel('Month')
plt.show()
```
