```python
#importing libraries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# loading datasets
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

#cleaning dataset
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = (df['job_skills']
                    .apply(lambda x: ast.literal_eval(x) 
                           if pd.notna(x) else x))
```

Plotting a pie chart

```python
df['job_work_from_home'].value_counts().plot(kind = 'pie', startangle = 90, autopct = '%1.1f%%')
plt.title('Work from Home Status')
plt.ylabel("")
plt.show()
```

```python
df['job_title_short'].value_counts().plot(kind = 'pie')
plt.title('Job Titles in the Dataset')
plt.ylabel("")
plt.show()
```

```python
df['job_title_short'].value_counts().plot(kind = 'barh')
plt.title('Job Titles in the Dataset')
plt.ylabel("")
plt.gca().invert_yaxis()
plt.show()
```

USING THE KNOWLEDGE OF SUBPLOTS

```python
fig, ax = plt.subplots(1, 3)

dict_col = {
    'job_work_from_home': 'Work from Home',
    'job_health_insurance': 'Health Insurance Offered',
    'job_no_degree_mention': 'Job Degree Req.'
}

for i, (column, title) in enumerate(dict_col.items()):
    ax[i].pie(df[column].value_counts(), startangle = 90, 
              autopct = '%1.1f%%', labels = ['False', 'True'])
    ax[i].set_title(title)

fig.tight_layout()

plt.show()
```
